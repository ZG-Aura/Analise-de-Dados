```python
import pandas as pd
```

```python
tabela = pd.read_excel("/content/drive/MyDrive/Data/Vendas.xlsx")
```

```python
tabela.head()
```

```python
#cálculo do faturamento total da empresa
faturamento_total = tabela["Valor Final"].sum()
faturamento_total
```

```python
#cálculo do faturamento por loja
faturamento_loja = tabela[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
faturamento_loja
```

```python
#cálculo do faturamento por loja e filtrando por produto
faturamento_produto = tabela[["ID Loja", "Produto", "Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
faturamento_produto
```
