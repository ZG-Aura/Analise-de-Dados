```python
import numpy as np
```

```python
def calculate(lista):
    if len(lista) != 9:
      raise ValueError("List must contain nine numbers.")

    matriz = np.array(lista).reshape(3,3)

    calculation = dict()
    calculation = {
        "mean": [matriz.mean(axis=0).tolist(), matriz.mean(axis=1).tolist(), matriz.mean()],
        "variance":[matriz.var(axis=0).tolist(), matriz.var(axis=1).tolist(), matriz.var()],
        "standard deviation":[matriz.std(axis=0).tolist(), matriz.std(axis=1).tolist(), matriz.std()],
        "max":[matriz.max(axis=0).tolist(), matriz.max(axis=1).tolist(), matriz.max()],
        "min":[matriz.min(axis=0).tolist(), matriz.min(axis=1).tolist(), matriz.min()],
        "sum":[matriz.sum(axis=0).tolist(), matriz.sum(axis=1).tolist(), matriz.sum()],
    }
    return calculation 
```

```python
calculate([0,1,2,3,4,5,6,7,8])
```
